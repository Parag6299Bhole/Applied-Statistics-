```python
from scipy.stats import ttest_ind
t_statistic, p_value  = ttest_ind(dc_new['Funds_in_million'].loc[dc_new['OperatingState']=='Operating'],dc_new['Funds_in_million'].loc[dc_new['OperatingState']=='Closed'])
print('tstat', t_statistic)    
print('P-Value', p_value)
```
